# Weather Forecasting

Installing all the packages for visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

Loading the dataset.

```python
data = pd.read_csv("DailyDelhiClimateTrain.csv")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

Changing the format of the date and time into specific format.

```python
data["date"] = pd.to_datetime(data["date"],format = '%Y-%m-%d')
```

```python
data.info()
```

Visualization for Mean Temperature.

```python
fig_meantemp = px.line(data,x = 'date',
                       y = "meantemp",
                       title = "Mean Temperature in Delhi")
fig_meantemp.show()
```

Visualization for humidity.

```python
fig_humidity = px.line(data,x = "date",
                       y = "humidity",
                       title = "Humidity in Delhi")
fig_humidity.show()
```

Visualization for wind speed.

```python
fig_wind_speed = px.line(data,x = "date",
                         y = "wind_speed",
                         title = "Wind Speed in Delhi")
fig_wind_speed.show()
```

Visualization for relatioship between temperature and humidity.

```python
figure = px.scatter(data_frame = data, x="humidity",
                    y="meantemp", size="meantemp", 
                    trendline="ols", 
                    title = "Relationship Between Temperature and Humidity")
figure.show()
```

```python
data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
print(data.head())
```

Visualization of the temperature change over the years.

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

```python
forecast_data = data.rename(columns = {"date": "ds", 
                                       "meantemp": "y"})
print(forecast_data)
```

Predicting the next mean temperature using facebook's prophet model

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
